```python
# Import necessary libraries
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Load your DataFrame
df = pd.read_csv('Airline_data.csv')

# Convert 'Year' to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Set 'Year' as the index of the dataframe
df.set_index('Year', inplace=True)

# Create a figure and a set of subplots
fig, ax1 = plt.subplots()

# Plot American Airlines profit
color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('American Airlines Profit (in Billion USD)', color=color)
lns1 = ax1.plot(df.index, df['American Airlines'], color=color, marker='o', label='American Airlines Profit')  # added label
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for crude oil price
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Crude Oil Price (USD per Barrel)', color=color)
lns2 = ax2.plot(df.index, df['Crude Oil'], color=color, marker='o', label='Crude Oil Price')  # added label
ax2.tick_params(axis='y', labelcolor=color)

# Combine legends
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)

# Modify x-axis to show every single year
ax1.xaxis.set_major_locator(mdates.YearLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv('Airline_data.csv')

# Calculate the correlation coefficient
correlation = df['Crude Oil'].corr(df['American Airlines'])
print(f"The correlation coefficient between the crude oil price and the profit of American Airlines is {correlation}")

# Calculate the slope, intercept, and r-squared value for the regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Crude Oil'], df['American Airlines'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.regplot(x='Crude Oil', y='American Airlines', data=df)

# Add titles and labels
plt.title('Correlation between Crude Oil Prices and American Airlines Profit')
plt.xlabel('Crude Oil Prices (USD per Barrel)')
plt.ylabel('American Airlines Profit (in Billion USD)')

# Annotate each point with its corresponding year
for i in range(df.shape[0]):
    plt.text(df['Crude Oil'].iloc[i], df['American Airlines'].iloc[i], df['Year'].iloc[i])

# Show the plot
plt.show()

# Print the slope, intercept, and r-squared value below the plot
print(f"Line equation: y = {slope:.2f}x + {intercept:.2f}")
print(f"R-squared value: {r_value**2:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Read the data from the CSV file
df = pd.read_csv('Airline_data.csv')

# List of airlines
airlines = ["Emirates Airlines", "American Airlines", "Air France", "All Nippon Airways", "Qantas Airways", "British Airways", "Ethiopian Airlines", "LATAM Airlines"]

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Iterate over each airline
for airline in airlines:
    # Plot the scatter plot with regression line
    sns.regplot(x='Crude Oil', y=airline, data=df, label=f'{airline} (corr={df["Crude Oil"].corr(df[airline]):.2f})', ax=ax)

# Add titles and labels
plt.title('Correlation between Crude Oil Prices and Airlines Profit')
plt.xlabel('Crude Oil Prices (USD per Barrel)')
plt.ylabel('Airlines Profit (in 10 Billion USD)')

# Add a legend outside the graph
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=4)

# Show the plot
plt.tight_layout()
plt.show()
```
