```python
# Import necessary libraries
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Load your DataFrame
project_airlines_data = pd.read_csv("project_airlines_data.csv")

df = pd.read_csv('project_airlines_data.csv')

# Convert 'Year' to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Set 'Year' as the index of the dataframe
df.set_index('Year', inplace=True)

# Create a figure and a set of subplots
fig, ax1 = plt.subplots()

# Plot Air France profit
color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Air France Profit (in Billion USD)', color=color)
lns1 = ax1.plot(df.index, df['Air France'], color=color, marker='o', label='Air France Profit')  # added label
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for crude oil price
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Crude Oil Price (USD per Barrel)', color=color)
lns2 = ax2.plot(df.index, df['Crude Oil'], color=color, marker='o', label='Crude Oil Price')  # added label
ax2.tick_params(axis='y', labelcolor=color)

# Combine legends
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)

# Modify x-axis to show every single year
ax1.xaxis.set_major_locator(mdates.YearLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv('project_airlines_data.csv')

# Calculate the correlation coefficient
correlation = df['Crude Oil'].corr(df['Air France'])
print(f"The correlation coefficient between the crude oil price and the profit of Air France is {correlation}")

# Calculate the slope, intercept, and r-squared value for the regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Crude Oil'], df['Air France'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.regplot(x='Crude Oil', y='Air France', data=df)

# Add titles and labels
plt.title('Correlation between Crude Oil Prices and Air France Profit')
plt.xlabel('Crude Oil Prices (USD per Barrel)')
plt.ylabel('Air France Profit (in Billion USD)')

# Annotate each point with its corresponding year
for i in range(df.shape[0]):
    plt.text(df['Crude Oil'].iloc[i], df['Air France'].iloc[i], df['Year'].iloc[i])

# Show the plot
plt.show()

# Print the slope, intercept, and r-squared value below the plot
print(f"Line equation: y = {slope:.2f}x + {intercept:.2f}")
print(f"R-squared value: {r_value**2:.2f}")
```
